```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
import os
```

### Topics to search 

```python
main_topic = 'legalization'
cannabis_topics = ['cannabis', 'marijuana', 'weed', 'legalizeit', 'decriminalize' ,'medical marijuana', 'medical cannabis', \
                   'cannabis policy']
other_topics = ['drug policy', 'drug legalization', 'drug decriminalization', 'psychedelic therapy', 'psilocybin therapy', \
                'MDMA therapy', 'gun control']
all_topics = [main_topic] + cannabis_topics + other_topics
```

```python
def complete_df_dates(df):
    max_date = max(df.date)
    min_date = min(df.date)
    if '01-01' in min_date and '12-31' in max_date:
        return True
    return False
```

```python
def date_reformat(date):
    return '-'.join(date.split(' ')[0].split('-')[1:])
```

```python
def fill_missing_df(topic):
    folder = './topics/' + topic + '/'
    for df_name in os.listdir(folder):
        df = pd.read_csv(folder + df_name)
        if complete_df_dates(df):
            print(f'{topic} {df_name} is complete df!')
            continue
        year = int(df_name.split('.')[0])
        max_date = max(df.date)
        min_date = min(df.date)
        if '12-31' in max_date:
            print(f'Download tweets to {topic}, df = {df_name}, until {date_reformat(min_date)}')
            create_tweets_df_by_topic(topic, start_year=year, end_year=year, until_date=date_reformat(min_date))
        elif '01-01' in min_date:
            print(f'Download tweets to {topic}, df = {df_name}, from {date_reformat(max_date)}')
            create_tweets_df_by_topic(topic, start_year=year, end_year=year, since_date=date_reformat(max_date))
```

```python
def create_tweets_df_by_topic(topic, batch_size=10000, start_year=2015, end_year=2022, since_date='01-01', until_date='01-01'):
    """
    Create a CSV file containing tweets related to a specified topic for each year in the specified range.
    The CSV files will be stored in a folder named after the topic under the 'topics' folder. 
    The function saves the tweets in batches to reduce memory usage.

    Args:
        topic (str): The topic for which tweets will be scraped.
        batch_size (int, optional): The number of tweets to process in each batch. Default is 10,000.
        start_year (int, optional): The start year of the range for which tweets will be scraped. Default is 2015.
        end_year (int, optional): The end year of the range for which tweets will be scraped. Default is 2022.
        since_date (str, optional): The start date (month-day) of the range for which tweets will be scraped. Default is '01-01'.
        until_date (str, optional): The end date (month-day) of the range for which tweets will be scraped. Default is '01-01'.

    The function performs the following steps:
        1. Checks if a 'topics' folder exists in the current directory. If not, it creates one.
        2. Checks if a folder for the specified topic exists in the 'topics' folder. If not, it creates one.
        3. Scrapes tweets related to the specified topic for each year in the specified range.
        4. Processes the scraped tweets in batches and appends them to the corresponding CSV file.
    """

    
    if 'topics' not in os.listdir('./'):
        os.mkdir('./topics')
        print('topics folder has been created!')
    if topic not in os.listdir('./topics/'):
        os.mkdir('./topics/' + topic)
        print(topic, 'folder has been created!')
        
    years = range(start_year, end_year + 1)
    
    for year in years:
        output_file = './topics/' + topic + '/' + str(year) + '.csv'
        tweets = []
        query = topic + ' until:' + str(year+1) + '-' + until_date + ' since:'  + str(year) + '-' + since_date
        header_written = False
        for i, tweet in enumerate(sntwitter.TwitterSearchScraper(query, maxEmptyPages=0).get_items()):
            tweets.append([
                tweet.date,
                tweet.user.username,
                tweet.rawContent,
                tweet.likeCount,
                tweet.replyCount,
                tweet.retweetCount,
                tweet.hashtags,
                tweet.id,
                tweet.user.id
            ])
            
            if len(tweets) % batch_size == 0 and len(tweets) != 0:
                print(f'year = {year}, iteration = {i+1}, date = {tweet.date}')
                if output_file.split('/')[-1] in os.listdir('./topics/' + topic):
                    header_written = True
                df = pd.DataFrame(tweets, columns=['date', 'user', 'tweet', 'likes', 'replies', 'retweets', 'hashtags', 'tweet_id', 'user_id'])
                with open(output_file, 'a' if header_written else 'w', encoding='utf-8') as f:
                    df.to_csv(f, index=False, header=not header_written)
                header_written = True
                tweets = []

        if tweets:  # save remaining tweets in the batch
            df = pd.DataFrame(tweets, columns=['date', 'user', 'tweet', 'likes', 'replies', 'retweets', 'hashtags', 'tweet_id', 'user_id'])
            with open(output_file, 'a' if header_written else 'w', encoding='utf-8') as f:
                df.to_csv(f, index=False, header=not header_written)

        print(f'year {year} done!')
```
